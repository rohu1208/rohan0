```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
plt.show()
```

```python
#plotting histogram
histplot(data=titanic_data,x='Fare',bins=30,kde=True,color='red')
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from Seaborn
titanic_df = sns.load_dataset('titanic')

# Plotting patterns in the data
plt.figure(figsize=(14, 7))

# Plotting survival rates based on passenger class
plt.subplot(2, 3, 1)
sns.countplot(data=titanic_df, x='pclass', hue='survived')
plt.title('Survival Counts by Passenger Class')

# Plotting survival rates based on gender
plt.subplot(2, 3, 2)
sns.countplot(data=titanic_df, x='sex', hue='survived')
plt.title('Survival Counts by Gender')

# Plotting survival rates based on embarkation port
plt.subplot(2, 3, 3)
sns.countplot(data=titanic_df, x='embarked', hue='survived')
plt.title('Survival Counts by Embarkation Port')

# Plotting survival rates based on number of siblings/spouses aboard
plt.subplot(2, 3, 4)
sns.countplot(data=titanic_df, x='sibsp', hue='survived')
plt.title('Survival Counts by Siblings/Spouses')

# Plotting survival rates based on number of parents/children aboard
plt.subplot(2, 3, 5)
sns.countplot(data=titanic_df, x='parch', hue='survived')
plt.title('Survival Counts by Parents/Children')

plt.tight_layout()
plt.show()
```
