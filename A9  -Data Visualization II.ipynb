1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution of 
age with respect to each gender along with the information about whether they survived or 
not. (Column names : 'sex' and 'age') 
2. Write observations on the inference from the above statistics.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import histplot
import pandas as pd
```

```python
titanic_data = pd.read_csv("C:\\Users\\rohan\\OneDrive\\Documents\\TE\\Data Science Lab\\A1-Data Wrangling1\\Titanic.csv")
```

```python
print(titanic_data.head())
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic_data, x='Sex', y='Age',color='pink', hue='Survived')
plt.title('Distribution of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

Observations:

1.The box plot reveals the distribution of ages for male and female passengers   separately.
2.We can observe any differences in age distribution between survivors (Yes)     and non-survivors (No) within each gender group.
3.From the box plot, we can observe any potential trends or differences in age   distribution among survivors and non-survivors for both genders.
4.This visualization helps us understand the relationship between gender, age,   and survival status, providing insights into potential factors influencing     survival on the Titanic.
